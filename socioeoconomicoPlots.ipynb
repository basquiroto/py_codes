```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Gráfico de Barras (e.g. População)

```python
pop = [{'ano': 2000, 'tipo': 'Rural', 'pop': 2549},
       {'ano': 2000, 'tipo': 'Urbana', 'pop': 18052},
       {'ano': 2000, 'tipo': 'Total', 'pop': 20601},
       {'ano': 2010, 'tipo': 'Rural', 'pop': 1637},
       {'ano': 2010, 'tipo': 'Urbana', 'pop': 23666},
       {'ano': 2010, 'tipo': 'Total', 'pop': 25303},
       {'ano': 2020, 'tipo': 'Rural', 'pop': None},
       {'ano': 2020, 'tipo': 'Urbana', 'pop': None},
       {'ano': 2020, 'tipo': 'Total', 'pop': 33611}]
 
pop_df = pd.DataFrame(pop)

sns.set_style('whitegrid')
fig = plt.gcf()
fig.set_size_inches(10, 6)

plot = sns.barplot(x = 'ano', y = 'pop', data = pop_df, hue = 'tipo')
plot.legend(title= '', fontsize = 14)

plot.set_yticklabels(plot.get_yticks().astype('int'), size = 12)

plt.xlabel('')
plt.ylabel('População', size = 12)

for p in plot.patches:
    plot.annotate("{:,.0f}".format(p.get_height()).replace(',','.'), (p.get_x() + p.get_width() / 2., p.get_height()), 
                  ha = 'center', va = 'center', xytext = (0, 8), textcoords = 'offset points', size = 13, weight = 'bold')
    
plt.savefig('C:\\Users\\ferna\\Desktop\\figura_Populacao.png',
            dpi = 120, bbox_inches='tight')
```

### Gráfico Pizza (e.g. Moradores)

```python
moradores = [{'1 morador': 15, '2 moradores': 50, '3 moradores': 36, 
              '4 moradores': 32, '5 moradores': 21, '6 ou mais': 25}]

mor_df = pd.DataFrame(moradores)

cores = sns.color_palette('pastel')[0:6]
dist = [0.4, 0, 0, 0, 0, 0]

plt.pie(mor_df.iloc[0], labels = mor_df.columns, colors = cores, autopct='%.0f%%',
       shadow = True, explode = dist)
plt.title('Quantidade de Moradores por Residência')

plt.savefig('C:\\Users\\ferna\\Desktop\\figura_Moradores.png',
            dpi = 120, bbox_inches='tight')
```

### Gráfico de linhas (e.g. Empregos)

```python
emp = [{'ano': 2010, 'Setor': 'Mineração', 'num': 2469},
       {'ano': 2010, 'Setor': 'Comércio', 'num': 3550},
       {'ano': 2010, 'Setor': 'Agricultura', 'num': 3037},
       {'ano': 2011, 'Setor': 'Mineração', 'num': 2589},
       {'ano': 2011, 'Setor': 'Comércio', 'num': 3690},
       {'ano': 2011, 'Setor': 'Agricultura', 'num': 3079},
       {'ano': 2012, 'Setor': 'Mineração', 'num': 2699},
       {'ano': 2012, 'Setor': 'Comércio', 'num': 3402},
       {'ano': 2012, 'Setor': 'Agricultura', 'num': 3335},
       {'ano': 2013, 'Setor': 'Mineração', 'num': 2607},
       {'ano': 2013, 'Setor': 'Comércio', 'num': 3511},
       {'ano': 2013, 'Setor': 'Agricultura', 'num': 3611},
       {'ano': 2014, 'Setor': 'Mineração', 'num': 2701},
       {'ano': 2014, 'Setor': 'Comércio', 'num': 3556},
       {'ano': 2014, 'Setor': 'Agricultura', 'num': 3778},
       {'ano': 2015, 'Setor': 'Mineração', 'num': 2753},
       {'ano': 2015, 'Setor': 'Comércio', 'num': 3570},
       {'ano': 2015, 'Setor': 'Agricultura', 'num': 3688}]

emp_df = pd.DataFrame(emp)
MEDIA = round(emp_df.num.mean(), 2)

sns.set_style('darkgrid')
grafico = sns.lineplot('ano', 'num', data = emp_df, hue = 'Setor',
                      marker='o')

# Remove título da legenda no Seaborn (i.e. Setor)
# https://tmlb-blog-ltd.medium.com/to-remove-a-title-in-the-legend-of-a-seaborn-graph-24c77345b5fb

handles0, labels0 = grafico.get_legend_handles_labels()
ggz = grafico.legend(handles=handles0[1:], labels=labels0[1:], loc='best', framealpha = 1,
                     fancybox=True, frameon=True).get_frame().set_facecolor('white')

plt.xlabel('Ano', size = 11)
plt.ylabel('Número de Empregados', size = 11)

grafico.axhline(MEDIA, color = 'r', linestyle = ':')
grafico.annotate('Média Geral', (2010, MEDIA*1.01))

plt.savefig('C:\\Users\\ferna\\Desktop\\figura_Emprego.png',
            dpi = 120, bbox_inches='tight')
```
