```python
import geobr as gbr
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Patch
from matplotlib.lines import Line2D
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Introdução ao geobr no Colab: https://www.youtube.com/playlist?list=PLR0bz-ML23oqeb4GkVBX550aDCZDUY41G
```

```python
Estados = gbr.read_state()
SC_BR = gbr.read_state(code_state='SC')
SC = gbr.read_municipality(code_muni='SC', year=2020)
Criciuma = gbr.read_municipality(code_muni=4204608, year=2020) # https://www.ibge.gov.br/explica/codigos-dos-municipios.php

# print(SC.columns, '\n', Estados.columns)
```

```python
#plt.rc({'font.size': 10})
tam_linha = 0.5

fig, ax = plt.subplots(figsize=(7, 7), dpi=120)

ax.add_patch(Rectangle((-51, -30), 4, 5,color="#AAFEFF", zorder = 1))
Estados.plot(facecolor="#FFFFFF", edgecolor="#000000", linewidth = tam_linha, ax = ax, zorder = 2)
SC.plot(facecolor="#FFFFFF", edgecolor="#000000", linewidth = tam_linha, ax = ax, zorder = 3)
Criciuma.plot(facecolor="#FF7A7A", edgecolor="#FF0000", linewidth = tam_linha+0.5, ax = ax, zorder = 4)

lages = plt.annotate("Lages", xy=(-50.26, -27.81), xycoords='data', xytext=(-51,-28.8), textcoords='data', 
                     arrowprops=dict(arrowstyle="-|>", color="#00FF00", lw=1))
parana = plt.annotate('Paraná', xy=(-51.5, -25.8), xycoords='data')
rgs = plt.annotate('Rio Grande do Sul', xy=(-52, -29.2), xycoords='data')
oceano = plt.annotate('Oceano\nAtlântico', xy=(-48.5, -29.2), xycoords='data', ha = 'center')
plt.setp(lages, size=12, color = '#00FF00')
plt.setp(parana, size = 8, color = '#777777')
plt.setp(rgs, size = 8, color = '#777777')
plt.setp(oceano, size = 7, color = '#888888')

#plt.yticks(rotation=90, va = 'center')
ax.tick_params(axis='both', which='major', pad=2, length = 2)

ax2 = fig.add_axes([0.12, 0.215, 0.25, 0.25]) # left, bottom, width, height %
ax2.tick_params(top=False, bottom=False, left=False, right=False, labelleft=False, labelbottom=False)
ax2.set_xlim(-75, -32)
ax2.set_ylim(-35, 7)
Estados.plot(facecolor="#FFFFFF", edgecolor="#000000", linewidth = tam_linha, ax = ax2, zorder = 5)
SC_BR.plot(facecolor="#FF0000", edgecolor="#000000", linewidth = tam_linha, ax = ax2, zorder = 6)

ax.set_title("Localização de Criciúma/SC", fontsize = 14)
ax.set_xlim([-54,-48])
ax.set_ylim([-29.5,-25.5])
ax.tick_params(axis='both', labelsize=8)
legenda = [Patch(facecolor="#FF7A7A", edgecolor="#FF0000", label='Criciúma'),
           Patch(facecolor="#FF0000", edgecolor="#000000", label='Santa Catarina')]
ax.legend(handles= legenda, loc = 'best')
```

```python
fig2, ax3 = plt.subplots(figsize=(7, 7), dpi=120)

SC.plot(facecolor="#FFFFFF", edgecolor="#000000", linewidth = tam_linha, ax = ax3, zorder = 1)
Criciuma.plot(facecolor="#FF7A7A", edgecolor="#FF0000", linewidth = tam_linha+0.5, ax = ax3, zorder = 2)

#empreend = plt.annotate("Empreendimento", xy=(-49.37, -28.68), xycoords='data', 
#                        xytext=(-49.32,-28.72), textcoords='data', 
#                        arrowprops=dict(arrowstyle="->", color="#000000", lw=1))
#plt.setp(empreend, size=12, color = '#000000')
ax3.add_patch(plt.Circle((-49.37, -28.68), .0025, facecolor="#FFFF55", 
                         edgecolor = "#000000", linewidth = tam_linha+1, zorder = 3))

#plt.yticks(rotation=90, va = 'center')
ax3.tick_params(axis='both', which='major', pad=2, length = 2)

ax4 = fig2.add_axes([0.135, 0.13, 0.25, 0.25]) # left, bottom, width, height %
ax4.tick_params(top=False, bottom=False, left=False, right=False, labelleft=False, labelbottom=False)
ax4.set_xlim(-54, -48)
ax4.set_ylim(-29.5, -25.5)

SC.plot(facecolor="#FFFFFF", edgecolor="#000000", linewidth = tam_linha, ax = ax4, zorder = 4)
Criciuma.plot(facecolor="#FF7A7A", edgecolor="#FF0000", linewidth = tam_linha+0.5, ax = ax4, zorder = 5)

ax3.set_title("Localização do Empreendimento", fontsize = 14)
ax3.set_xlim([-49.55,-49.2])
ax3.set_ylim([-28.88,-28.6])
ax3.tick_params(axis='both', labelsize=8)
legenda = [Line2D([0], [0], marker='o', color='w', markerfacecolor='#FFFF55', 
                  markeredgecolor= '#000000', markersize=8, markeredgewidth=tam_linha+1,
                  label = 'Empreendimento'), 
           Patch(facecolor="#FF7A7A", edgecolor="#FF0000", label='Criciúma')]
ax3.legend(handles= legenda, loc = 'best')
```
